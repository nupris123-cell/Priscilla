```python
# Task Descriptive Statistics Questions
# 2. Age Analysis
# Part 1

# Read file and load file to dataframe

import pandas as pd

file_path ="Customer Satisfaction.csv"

try:
    dfCustomerSatisfaction = pd.read_csv(file_path)
    print("CSV file loaded successfully into a DataFrame.")
    display(dfCustomerSatisfaction.head())
except FileNotFoundError:
    print(f"Error: The file '{file_path}' was not found.")
except Exception as e:
    print(f"An error occurred: {e}")

# Average Age in Customer Satisfaction dataframe

avg_age = dfCustomerSatisfaction["Age"].mean()
print(f"Average age: {avg_age:.2f}")
```

```python
# Part 2

import pandas as pd
import matplotlib.pyplot as plt
import sqlite3
import numpy as np

# Create SQLite in-memory database

conn = sqlite3.connect(':memory:')

# Store DataFrame as a SQL table

dfCustomerSatisfaction.to_sql('CustomerSatisfaction', conn, index=False, if_exists='replace')


# Query in SQL

query = "SELECT MIN(Age) FROM CustomerSatisfaction"

min_age = pd.read_sql_query(query, conn).iloc[0, 0]
print("Minimum age:", min_age)

query = "SELECT MAX(Age) FROM CustomerSatisfaction"

max_age = pd.read_sql_query(query, conn).iloc[0, 0]
print("Maximum age:", max_age)

print("Age Range:", max_age - min_age)

# Plot box plot for Age Range

plt.boxplot(dfCustomerSatisfaction["Age"])
plt.ylabel("Age")
plt.title("Age Range")
plt.show()

# Count of travellers below 18 years age

query = "SELECT COUNT(*) FROM CustomerSatisfaction WHERE Age < 18"
Ageless18 = pd.read_sql_query(query, conn).iloc[0, 0]
print("Count of travellers below 18 years age:", Ageless18)

# Count of travellers between 18-30 years age

query = "SELECT COUNT(*) FROM CustomerSatisfaction WHERE Age >= 18 AND Age <= 30"
Age18_30 = pd.read_sql_query(query, conn).iloc[0, 0]
print("Count of travellers between 18-30 years age:", Age18_30)

# Count of travellers between 31-45 years age

query = "SELECT COUNT(*) FROM CustomerSatisfaction WHERE Age >= 31 AND Age <= 45"
Age31_45 = pd.read_sql_query(query, conn).iloc[0, 0]
print("Count of travellers between 31-45 years age:", Age31_45)

# count of travellers between 46-60 years age

query = "SELECT COUNT(*) FROM CustomerSatisfaction WHERE Age >= 46 AND Age <= 60"
Age46_60 = pd.read_sql_query(query, conn).iloc[0, 0]
print("Count of travellers between 46-60 years age:", Age46_60)

# count of travellers above 60 years age

query = "SELECT COUNT(*) FROM CustomerSatisfaction WHERE Age > 60"
Ageabove60 = pd.read_sql_query(query, conn).iloc[0, 0]
print("Count of travellers above 60 years age:", Ageabove60)

Total = Ageless18 + Age18_30 + Age31_45 + Age46_60 + Ageabove60
print("Total Count:", Total)
describe = dfCustomerSatisfaction["Age"].describe()
print(describe)

# plot a histogram with Age bracket distribution

labels = ['<18', '18-30', '31-45', '46-60', '>60']
counts = [Ageless18, Age18_30, Age31_45, Age46_60, Ageabove60]
colors = ['blue', 'green', 'red', 'purple', 'orange']
plt.bar(labels, counts, color=colors)
plt.xlabel('Age Bracket')
plt.ylabel('Count')
plt.title('Age Bracket Distribution')
plt.show()
```

```python
# Task Descriptive Statistics Questions
# 2. Travel Category
# Part 1

# Read file and load file to dataframe

import pandas as pd

file_path ="Customer Satisfaction.csv"

try:
    dfCustomerSatisfaction = pd.read_csv(file_path)
    print("CSV file loaded successfully into a DataFrame.")
    display(dfCustomerSatisfaction.head())
except FileNotFoundError:
    print(f"Error: The file '{file_path}' was not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python

# count of business vs Personal Travellers

import pandas as pd
import matplotlib.pyplot as plt
import sqlite3
import numpy as np

# Create SQLite in-memory database

conn = sqlite3.connect(':memory:')

# Store DataFrame as a SQL table

dfCustomerSatisfaction.to_sql('CustomerSatisfaction', conn, index=False, if_exists='replace')


# Query in SQL for finding number of trvellers for personal or official businee

query = "SELECT COUNT(*) FROM CustomerSatisfaction WHERE [Travel Category] = 'Personal Travel'"

PersonalTravelers = pd.read_sql_query(query, conn).iloc[0, 0]
print("Personal Travelers:", PersonalTravelers)

query = "SELECT COUNT(*) FROM CustomerSatisfaction WHERE [Travel Category] = 'Official'"

OfficialTravelers = pd.read_sql_query(query, conn).iloc[0, 0]
print("Business Travelers:", OfficialTravelers)
# Bar chart to show number of travelere in each Travel Category

labels = ['Personal Travel', 'Official']
counts = [PersonalTravelers, OfficialTravelers]
colors = ['orange', 'yellow']
plt.bar(labels, counts, color=colors)
plt.xlabel('Travel Category')
plt.ylabel('Count')
plt.title('Travel Category Distribution')
plt.show()
```

```python
# Part 2

# Average distance traveled category wise

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Using pandas query to find average

grouped = dfCustomerSatisfaction.groupby("Travel Category")["Distance Travelled"].mean()
print(grouped)

# box plot average distance traveled travel category wise

grouped.plot(kind="bar",color='cyan')
plt.xlabel("Travel Category")
plt.ylabel("Average Distance Travelled")
plt.title("Average Distance Travelled by Travel Category")
plt.show()
```

```python
# 4. Travel Class Ratings
# Part 1
# Average rating of seat comfort across different travel classes
# check for missing values in seat comfort column

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


# print(dfCustomerSatisfaction['Seat Comfort'].isnull())        # Boolean missing values
print(dfCustomerSatisfaction['Seat Comfort'].isnull().sum())  # Count missing values per column


# Fill with column median

seat_comfort_median = dfCustomerSatisfaction["Seat Comfort"].median()
dfCustomerSatisfaction["Seat Comfort"] = dfCustomerSatisfaction["Seat Comfort"].fillna(seat_comfort_median)
print(f"Median seat comfort used for filling: {seat_comfort_median}")
print(dfCustomerSatisfaction['Seat Comfort'].isnull().sum())  # Count missing values per column

# Average seat comfort raring across categoryTravel class

grouped = dfCustomerSatisfaction.groupby("Travel Class")["Seat Comfort"].mean()
print(grouped)

# Plot bar chart for Seat Comfort Rating across Travel Class

grouped.plot(kind="bar", color=['blue','red','green'])
plt.xlabel("Travel Class")
plt.ylabel("Average Seat Comfort Rating")
plt.title("Average Seat Comfort Rating by Travel Class")
plt.show()
```

```python
# Part 2
# Rating difference for food between business and economy class

# check for missing values in seat comfort column

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


# print(dfCustomerSatisfaction['Food'].isnull())        # Boolean missing values
print(dfCustomerSatisfaction['Food'].isnull().sum())  # Count missing values per column


# Fill with column median

Food_median = dfCustomerSatisfaction["Food"].median()
dfCustomerSatisfaction["Food"] = dfCustomerSatisfaction["Food"].fillna(Food_median)
print(f"Median Food Rating used for filling: {Food_median}")
print(dfCustomerSatisfaction['Food'].isnull().sum())  # Count missing values per column

# Average Food rating across categoryTravel class

grouped = dfCustomerSatisfaction.groupby("Travel Class")["Food"].mean()
print(grouped)

# Plot bar chart for Seat Comfort Rating for Travel Class 'Business' vs 'Economy'

# choose which categories to plot
to_plot = grouped.loc[['Business', 'Economy']]   # or grouped.drop('Premium')

to_plot.plot(kind="bar", color=['pink','yellow'])
plt.xlabel("Travel Class")
plt.ylabel("Average Food Rating")
plt.title("Average Food Rating by Travel ClassBusiness vs. Economy")
plt.show()
```

```python
# 5. Delay Analysis
# Part 1
# Average departure and arrival delay

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# No relevant null values in dataset!

DepartureDelay = dfCustomerSatisfaction["Departure Delay (min)"].mean()
ArrivalDelay = dfCustomerSatisfaction["Arrival Delay (min)"].mean()

print(f"Average Departure Delay: {DepartureDelay}")
print(f"Average Arrival Delay: {ArrivalDelay}")

# Plot bar chart to show average Departure and Arrival delays

labels = ['Departure Delay', 'Arrival Delay']
Delays = [DepartureDelay, ArrivalDelay]
colors = ['yellow', 'orange']
plt.bar(labels, Delays, color=colors)
plt.xlabel('Delay Type')
plt.ylabel('Average Delay in minutes')
plt.title('Average Departure and Arrival Delays')
plt.show()
```

```python
# Part 2
# Impact of delays both departure and arrival on customer Satisfaction

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Add a new column, Total Delay in the dataset, concatenate

dfCustomerSatisfaction["Total Delay"] = dfCustomerSatisfaction["Departure Delay (min)"] + dfCustomerSatisfaction["Arrival Delay (min)"]

# Fill NaN values with 0 before converting to integer
dfCustomerSatisfaction["Total Delay"] = dfCustomerSatisfaction["Total Delay"].fillna(0).astype(int)

# print result and plot stacked bar graph to show Satisfaction result with delay and with no delay

# Create a new column for delay status
dfCustomerSatisfaction["Delay Status"] = dfCustomerSatisfaction["Total Delay"].apply(lambda x: "With Delay" if x > 0 else "Without Delay")

# Count Satisfaction per Delay Status
counts = dfCustomerSatisfaction.groupby(["Delay Status", "Satisfaction"]).size().unstack(fill_value=0)

print(counts)
# Plot stacked bar chart
counts.plot(kind="bar", stacked=True, figsize=(8, 6), color=['pink','cyan'])

plt.title("Customer Satisfaction With vs Without Delay")
plt.ylabel("Number of Customers")
plt.xlabel("Delay Status")
plt.legend(title="Satisfaction")
plt.show()
```

AS we can see from above, the number of travelers who are dissatisfied are slightly higher when there is no delay as compared to when there is a delay.
